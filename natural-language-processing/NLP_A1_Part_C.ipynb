# Add to the pipeline a custom component that prints the longest token in the document

```python
@Language.component('custom_component')
def custom_component_function(doc):
    # print the longest token of the document
    longest_token = max(doc, key=len)
    print('\nThe longest token of this document is: {}\nLength: {}'.format(longest_token, len(longest_token)))

    return doc
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('custom_component', first=True)

for component in nlp.pipeline:
    print(component)

doc = nlp('min medium maximum')
```
